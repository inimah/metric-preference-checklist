### For each human-like aspect, Compute human scores for all models 

The resulting score is used to pair model based on their level of difficulty

Difficult pair = the score gap between two models is not significantly discernable

Easy pair = two models have a significant score difference

```python
# get the pair list of the model's name to be further used to compute KS score for each evaluation metrics
# including human metrics
i=0
data_model_pairs = {}
for dat in list(dfs.keys()):
    df = dfs[dat]
    if(i==1):
        df['overall'] = (df['coherence'] + df['consistency'] + df['fluency'] + df['relevance'])/4
    elif(i==2):
        df['overall'] = (df['CoherenceRating'] + \
                         df['FluencyRating'] + df['InformativenessRating'] + df['RelevanceRating'])/4
    print(dat)
    #print(df.columns)
    model_col_name = model_naming[i][0]
    #print(model_col_name)
    
    model_pairs = {} 
    easy_pairs = [] # array of tuple
    hard_pairs = []
    for humm in hum_metr[i]:
        print(humm)
        hum_sc = {}
        for m in np.unique(df[model_col_name].values):
            df_model = df[df[model_col_name]==m]
            score = df_model[humm].values
            hum_sc[m] = np.mean(score)
        sorted_hum_sc = {k: v for k, v in sorted(hum_sc.items(), key=lambda item: item[1])}
        for k,v in sorted_hum_sc.items():
            print(k,v)
        # easy differentiate = low-Q vs high-Q system
        lenid =  len(sorted_hum_sc.items()) -1
        midid = int(len(sorted_hum_sc.items())/2)
        # only use this method for dataset containing > 4 models/systems
        # else manually select
        if len(np.unique(df[model_col_name].values)) > 4:
            pairs1 = tuple([list(sorted_hum_sc.keys())[0], list(sorted_hum_sc.keys())[lenid]])
            pairs2 = tuple([list(sorted_hum_sc.keys())[0], list(sorted_hum_sc.keys())[lenid-1]])
            pairs3 = tuple([list(sorted_hum_sc.keys())[1], list(sorted_hum_sc.keys())[lenid]])
            pairs4 = tuple([list(sorted_hum_sc.keys())[1], list(sorted_hum_sc.keys())[lenid-1]])
            if pairs1 not in easy_pairs:
                easy_pairs.append(pairs1)
            if pairs2 not in easy_pairs:
                easy_pairs.append(pairs2)
            if pairs3 not in easy_pairs:
                easy_pairs.append(pairs3)
            if pairs4 not in easy_pairs:
                easy_pairs.append(pairs4)
        # only use this method for dataset containing > 6 models/systems
        # else manually select
        if len(np.unique(df[model_col_name].values)) > 6:
            # hard-1 = low-Q vs low-Q
            hard1 = tuple([list(sorted_hum_sc.keys())[0], list(sorted_hum_sc.keys())[1]])
            # hard-2 = mod-Q vs. mod-Q
            hard2 = tuple([list(sorted_hum_sc.keys())[midid], list(sorted_hum_sc.keys())[midid+1]])
            # hard-3 = high-Q vs high-Q
            hard3 = tuple([list(sorted_hum_sc.keys())[lenid-1], list(sorted_hum_sc.keys())[lenid]])
            if hard1 not in hard_pairs:
                hard_pairs.append(hard1)
            if hard2 not in hard_pairs:
                hard_pairs.append(hard2)
            if hard3 not in hard_pairs:
                hard_pairs.append(hard3)            
        print()
    if len(easy_pairs) != 0:
        model_pairs['easy'] = easy_pairs
    if len(hard_pairs) != 0:
        model_pairs['hard'] = hard_pairs
    data_model_pairs[dat] = model_pairs
    print('-----')
    i+=1
```

## Compute KS score

```python
from scipy import stats
```

```python
i=0
data_ks_score_easy = {}
data_ks_score_hard = {}
data_sys_prefer_easy = {}
data_sys_prefer_hard = {}

for k,v in data_model_pairs.items():
    print(k)
    # get easy tuples
    easy_pairs = v['easy']
    # get hard tuples
    hard_pairs = v['hard']
    df = dfs[k] # get dataset
    mod_name = model_naming[i][0]
    ks_score_easy = {}
    ks_score_hard = {}
    sys_prefer_easy = {}
    sys_prefer_hard = {}
    # KS of human metriccs
    for humm in hum_metr[i]:
        ks_scores1 = []
        p_scores1 = []
        ks_scores2 = []
        p_scores2 = []
        pref1 = []
        pref2 = []
        for easy_ in easy_pairs:
            
            df1 = df[df[mod_name]==easy_[0]]
            df2 = df[df[mod_name]==easy_[1]]
            sample1 = df1[humm].values
            sample2 = df2[humm].values
            if np.mean(sample1) < np.mean(sample2):
                #pref = str(easy_[0]) + "<" + str(easy_[1])
                pref =  "a<b" 
            elif np.mean(sample1) > np.mean(sample2):
                #pref = str(easy_[1]) + "<" + str(easy_[0])
                pref =  "b<a"
            else:
                #pref = str(easy_[0]) + "=" + str(easy_[1])
                pref =  "a=b"
            if k=='newsroom_unieval_summ.csv' and 'CoherenceRating':
                print("================")
                print(k)
                print(easy_)
                print(np.mean(sample1))
                print(np.mean(sample2))
                print(pref)
                print("================")
            pref1.append(pref)
            ks_score = stats.ks_2samp(sample1, sample2) # ks_score, p-significant-val
            #score = tuple([ks_score[0], ks_score[1]])
            ks_scores1.append(ks_score[0])
            p_scores1.append(ks_score[1])
        #id_ = np.argmax(ks_scores1)
        #mean_ks_scores1 = ks_scores1[id_]
        #mean_p_scores1 = p_scores1[id_]
        mean_ks_scores1 = np.mean(ks_scores1)
        #mean_p_scores1 = np.mean(p_scores1)
        #ks_score_easy[humm] = tuple([mean_ks_scores1, mean_p_scores1])
        ks_score_easy[humm] = mean_ks_scores1
        sys_prefer_easy[humm] = pref1
        for hard_ in hard_pairs:
            # KS of human metriccs
            df1 = df[df[mod_name]==hard_[0]]
            df2 = df[df[mod_name]==hard_[1]]
            sample1 = df1[humm].values
            sample2 = df2[humm].values
            if np.mean(sample1) < np.mean(sample2):
                #pref = str(easy_[0]) + "<" + str(easy_[1])
                pref =  "a<b" 
            elif np.mean(sample1) > np.mean(sample2):
                #pref = str(easy_[1]) + "<" + str(easy_[0])
                pref =  "b<a"
            else:
                #pref = str(easy_[0]) + "=" + str(easy_[1])
                pref =  "a=b"
            pref2.append(pref)
            ks_score = stats.ks_2samp(sample1, sample2) # ks_score, p-significant-val
            #score = tuple([ks_score[0], ks_score[1]])
            ks_scores2.append(ks_score[0])
            p_scores2.append(ks_score[1])
        #id_ = np.argmax(ks_scores2)   
        #mean_ks_scores2 = ks_scores2[id_]
        #mean_p_scores2 = p_scores2[id_]
        mean_ks_scores2 = np.mean(ks_scores2)
        #mean_p_scores2 = np.mean(p_scores2)
        #ks_score_hard[humm] = tuple([mean_ks_scores2, mean_p_scores2])
        ks_score_hard[humm] = mean_ks_scores2
        sys_prefer_hard[humm] = pref2
    # KS of automated metrics
    for automm in autom_metr[i]:
        print(automm)
        ks_scores1 = []
        p_scores1 = []
        ks_scores2 = []
        p_scores2 = []
        pref1 = []
        pref2 = []
        for easy_ in easy_pairs:
            df1 = df[df[mod_name]==easy_[0]]
            df2 = df[df[mod_name]==easy_[1]]
            sample1 = df1[automm].values
            sample2 = df2[automm].values
            if np.mean(sample1) < np.mean(sample2):
                #pref = str(easy_[0]) + "<" + str(easy_[1])
                if automm != "perplexity":
                    pref =  "a<b" 
                else:
                    pref =  "b<a" 
            elif np.mean(sample1) > np.mean(sample2):
                #pref = str(easy_[1]) + "<" + str(easy_[0])
                if automm != "perplexity":
                    pref =  "b<a"
                else:
                    pref =  "a<b"
            else:
                #pref = str(easy_[0]) + "=" + str(easy_[1])
                pref =  "a=b"
            pref1.append(pref)
            ks_score = stats.ks_2samp(sample1, sample2) # ks_score, p-significant-val
            #score = tuple([ks_score[0], ks_score[1]])
            ks_scores1.append(ks_score[0])
            p_scores1.append(ks_score[1])
        #id_ = np.argmax(ks_scores1)
        #mean_ks_scores1 = ks_scores1[id_]
        #mean_p_scores1 = p_scores1[id_]
        mean_ks_scores1 = np.mean(ks_scores1)
        #mean_p_scores1 = np.mean(p_scores1)
        #ks_score_easy[automm] = tuple([mean_ks_scores1, mean_p_scores1])
        ks_score_easy[automm] = mean_ks_scores1
        sys_prefer_easy[automm] = pref1
        for hard_ in hard_pairs:
            # KS of human metriccs
            df1 = df[df[mod_name]==hard_[0]]
            df2 = df[df[mod_name]==hard_[1]]
            sample1 = df1[automm].values
            sample2 = df2[automm].values
            if np.mean(sample1) < np.mean(sample2):
                #pref = str(easy_[0]) + "<" + str(easy_[1])
                if automm != "perplexity":
                    pref =  "a<b" 
                else:
                    pref =  "b<a"
            elif np.mean(sample1) > np.mean(sample2):
                #pref = str(easy_[1]) + "<" + str(easy_[0])
                if automm != "perplexity":
                    pref =  "b<a"
                else:
                    pref =  "a<b"
            else:
                #pref = str(easy_[0]) + "=" + str(easy_[1])
                pref =  "a=b"
            pref2.append(pref)
            ks_score = stats.ks_2samp(sample1, sample2) # ks_score, p-significant-val
            #score = tuple([ks_score[0], ks_score[1]])
            ks_scores2.append(ks_score[0])
            p_scores2.append(ks_score[1])
        #id_ = np.argmax(ks_scores2) 
        #mean_ks_scores2 = ks_scores2[id_]
        #mean_p_scores2 = p_scores2[id_]
        mean_ks_scores2 = np.mean(ks_scores2)
        #mean_p_scores2 = np.mean(p_scores2)
        #ks_score_hard[automm] = tuple([mean_ks_scores2, mean_p_scores2]) 
        ks_score_hard[automm] = mean_ks_scores2
        sys_prefer_hard[automm] = pref2
    data_ks_score_easy[k] = ks_score_easy
    data_ks_score_hard[k] = ks_score_hard
    data_sys_prefer_easy[k] = sys_prefer_easy
    data_sys_prefer_hard[k] = sys_prefer_hard
    i+=1
```

### sim = (L1_correct + L2)-LevD/(L1+L2)

```python
from Levenshtein import *
from Levenshtein import distance as lev
```

```python
# function to calculate the similarity between preference pairs of automated metrics and human metrics
def compute_sim_preference(pref_automm, pref_humm):
    # both pref_automm and pref_humm are string
    lev_dist = lev(pref_automm, pref_humm)
    sim = (len(pref_automm) + len(pref_humm) - (2*lev_dist) )/(len(pref_automm) + len(pref_humm))
    return sim
    
```

## Preference similarity between human aspects 

```python
i=0
dat_scores_easy = {}
dat_scores_hard = {}
for k in dfs.keys():
    print(k)
    df_easy = data_sys_prefer_easy[k]
    df_hard = data_sys_prefer_hard[k]
    met_scores = {}
    met_scores_hard = {}
    for automm in autom_metr[i]:
        # get preference of autom metrics
        pref_autom_all = df_easy[automm]
        pref_autom_hard = df_hard[automm]
        hum_scores = {}
        hum_scores_hard = {}
        for humm in hum_metr[i]:
            # get preference of human metrics
            pref_humm_all = df_easy[humm]
            pref_humm_hard = df_hard[humm]
            scores = []
            for p1, p2 in zip(pref_autom_all, pref_humm_all):
                sim = compute_sim_preference(p1, p2)
                scores.append(sim)
            hum_scores[humm] = np.mean(scores)
            scores2 = []
            for p1, p2 in zip(pref_autom_hard, pref_humm_hard):
                sim = compute_sim_preference(p1, p2)
                scores2.append(sim)
            hum_scores_hard[humm] = np.mean(scores2)
        met_scores[automm] = hum_scores
        met_scores_hard[automm] = hum_scores_hard
    dat_scores_easy[k] = met_scores
    dat_scores_hard[k] = met_scores_hard
    i+=1
    
```
